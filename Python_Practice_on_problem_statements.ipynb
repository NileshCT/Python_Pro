```python
#Python Program to Find HCF or GCD
def findhcf(x,y):
  if x>y:
    smaller=y
  else:
    smaller=x
  for i in range(1,smaller+1):
    if (x%i==0) and (y%i==0):
          hcf=i
  return hcf
print("hcf is",findhcf(12,36))
```
